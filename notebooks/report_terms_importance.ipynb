### this is original Syed code

reads `terms_topics` from syed's report sheet associating each term to a topic with a relative weight (and also reporting on idf)

read list of terms

writes a new df with idf and rank position of each of the terms in each of the topics


```python
import pandas as pd


xls = pd.ExcelFile('final_x_idf.xlsx')
terms_topics = pd.read_excel(xls)

terms = pd.read_excel(xls,'terms')
```

```python
terms_topics
```

```python
def term2idf(term):
    return terms_topics[terms_topics.MLTC==term].idf.item()

def get_position(term, topic):
    terms_topics_sorted = terms_topics.sort_values(by=topic, ignore_index=True, ascending=False)
    return terms_topics_sorted[terms_topics_sorted.MLTC == term].index.values.astype(int)[0]
    
```
